```python
import pandas as pd 
import numpy as np

import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

from sklearn.model_selection import cross_val_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

from sklearn.preprocessing import FunctionTransformer
from sklearn.compose import ColumnTransformer

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('placementdata.csv' , usecols = ['CGPA' , 'AptitudeTestScore' , 'PlacementStatus'])
```

```python
df.CGPA.fillna(df.CGPA.mean() , inplace = True)
```

```python
df
```

```python
x = df.iloc[:, [0, 1]]
y = df.iloc[:, 2]  
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
plt.figure(figsize = (14,4))
plt.subplot(121)
sns.distplot(x_train.CGPA)
plt.title('CGPA Pdf')

plt.subplot(122)
stats.probplot(x_train.CGPA , dist = 'norm' , plot = plt)
plt.title('CGPA QQ plot')
```

```python
plt.figure(figsize = (14,4))
plt.subplot(121)
sns.distplot(x_train.AptitudeTestScore)
plt.title('AptitudeTestScore')

plt.subplot(122)
stats.probplot(x_train.AptitudeTestScore , dist = 'norm' , plot = plt)
plt.title('AptitudeTestScore QQ plot')
```

```python
clf = LogisticRegression()
clf1 = DecisionTreeClassifier()
```

```python
clf.fit(x_train , y_train)
clf1.fit(x_train , y_train)

y_pred = clf.predict(x_test)
y_pred1 = clf1.predict(x_test)

print('accuracy of lr' , accuracy_score(y_test , y_pred))
print('accuracy of dtr' , accuracy_score(y_test , y_pred1))
```

```python
trf = FunctionTransformer(func = np.log1p)
```

```python
x_train_transformed = trf.fit_transform(x_train)
x_test_transformed =  trf.transform(x_test)
```

```python
clf = LogisticRegression()
clf1 = DecisionTreeClassifier()

clf.fit(x_train_transformed , y_train)
clf1.fit(x_train_transformed , y_train)

y_pred = clf.predict(x_test_transformed)
y_pred1 = clf1.predict(x_test_transformed)

print('accuracy of lr' , accuracy_score(y_test , y_pred))
print('accuracy of dtc' , accuracy_score(y_test , y_pred1))
```

```python
x_transformed = trf.fit_transform(x)

clf = LogisticRegression()
clf1 = DecisionTreeClassifier()

print('LR' , np.mean(cross_val_score(clf , x_transformed , y , scoring = "accuracy" , cv = 10)))
print('DTC' , np.mean(cross_val_score(clf1 , x_transformed , y , scoring = "accuracy" , cv = 10)))
```

```python
plt.figure(figsize = (14,4))
plt.subplot(121)
stats.probplot(x_train.AptitudeTestScore , dist = 'norm' , plot = plt)
plt.title('AptitudeTestScore before log')

plt.subplot(122)
stats.probplot(x_train_transformed.AptitudeTestScore , dist = 'norm' , plot = plt)
plt.title('AptitudeTestScore after log')
```

```python
x=df.iloc[:,:-1]
y=df.PlacementStatus
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
x_train , x_test , y_train , y_test = train_test_split(x ,y, test_size = .25 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier  = LogisticRegression()
```

```python
parameter = {"penalty":["l1" ,"l2" , "elasticnet"] , "C":[1,2,3,4,56,5,7,9,30,54,60],"max_iter":[100,200,300,400,500]}
```

```python
classifier_reg = GridSearchCV(classifier , param_grid = parameter , scoring = "accuracy" , cv = 10)
```

```python
classifier_reg.fit(x_train , y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
classifier_reg.best_score_
```

```python
y_pred = classifier_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_pred , y_test)
```

```python
from sklearn.model_selection import cross_val_score

cv_scores = cross_val_score(classifier_reg.best_estimator_, x, y, cv=10, scoring="accuracy")
print("Cross-Validation Scores:", cv_scores)
print("Mean CV Accuracy:", np.mean(cv_scores))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df=pd.read_csv('placementdata.csv')
```

```python
df.isnull().sum()
```

```python
df.ExtracurricularActivities = df.ExtracurricularActivities.map({"No":0 , "Yes":1 })
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df.PlacementStatus
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test  = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(max_depth=2)
treemodel.fit(x_train , y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (20,15))
tree.plot_tree(treemodel, filled = True)
plt.tight_layout()
```

```python
tree_model = DecisionTreeClassifier(max_depth = 2)
tree_model.fit(x_train , y_train)
```

```python
tree.plot_tree(tree_model , filled = True)
plt.tight_layout()
```

```python
y1 = treemodel.predict(x_test)
y2 = tree_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
print(classification_report(y1 , y_test))
```

```python
print(classification_report(y2 , y_test))
```

```python
print(accuracy_score(y1 , y_test))
print(accuracy_score(y2 , y_test))

```
