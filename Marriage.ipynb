```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('marriage_data_india.csv')
```

```python
df
```

```python
df.Marriage_Type = df.Marriage_Type.map({"Love":0 , "Arranged":1 })
```

```python
df.Gender = df.Gender.map({"Male":0 , "Female":1 })
df.Education_Level = df.Education_Level.map({"School":0 , "Graduate":1, "Postgraduate":2, "PhD":3 })
```

```python
df.Caste_Match = df.Caste_Match.map({"Different":0 , "Same":1 })
df.Religion = df.Religion.map({"Hindu":0 , "Muslim":1 ,"Christian":2,"Sikh":3, "Others":4})
```

```python
import pandas as pd

# Replace categorical values with numerical values
df["Dowry_Exchanged"] = df["Dowry_Exchanged"].replace({
    "Yes": 1, 
    "No": 0, 
    "Not Disclosed": -1  
})


df["Dowry_Exchanged"].fillna(-1, inplace=True)  


df["Dowry_Exchanged"] = df["Dowry_Exchanged"].astype(int)


print(df["Dowry_Exchanged"].head())
```

```python
df.Parental_Approval = df.Parental_Approval.map({"No":0 , "Yes":1,"Partial":2 })
df.Urban_Rural = df.Urban_Rural.map({"Urban":0 , "Rural":1 })
df.Marital_Satisfaction = df.Marital_Satisfaction.map({"Low":0 , "High":1,"Medium":2 })
df.Divorce_Status = df.Divorce_Status.map({"No":0 , "Yes":1 })
df.Spouse_Working = df.Spouse_Working.map({"No":0 , "Yes":1 })
df['Inter-Caste'] = df['Inter-Caste'].map({"No":0 , "Yes":1 })
df['Inter-Religion'] = df['Inter-Religion'].map({"No":0 , "Yes":1 })
```

```python
df.Income_Level = df.Income_Level.map({"Low":0 , "Middle":1,"High":2 })
```

```python
df
```

```python
x = df.drop(columns=['Marriage_Type'])  # Independent Variables
y = df['Marriage_Type']
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier  = LogisticRegression()
```

```python
parameter = {"penalty":["l1" ,"l2" , "elasticnet"] , "C":[1,2,3,4,56,5,7,9,30,54,60],"max_iter":[100,200,300,400,500]}
```

```python
classifier_reg = GridSearchCV(classifier , param_grid = parameter , scoring = "accuracy" , cv = 10)
```

```python
classifier_reg.fit(x_train , y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
classifier_reg.best_score_
```

```python
y_pred = classifier_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_pred , y_test)
```

```python
print(classification_report(y_pred ,y_test))
```

```python
df.Caste_Match = df.Caste_Match.map({"Different":0 , "Same":1 })
df.Religion = df.Religion.map({"Hindu":0 , "Muslim":1 ,"Christian":2,"Sikh":3, "Others":4})
```

```python
import pandas as pd

# Replace categorical values with numerical values
df["Dowry_Exchanged"] = df["Dowry_Exchanged"].replace({
    "Yes": 1, 
    "No": 0, 
    "Not Disclosed": -1  # You can choose -1, or handle it differently
})

# Handle NaN values before type conversion
df["Dowry_Exchanged"].fillna(-1, inplace=True)  # Replace NaN with -1 (or another strategy)

# Convert column to integer type
df["Dowry_Exchanged"] = df["Dowry_Exchanged"].astype(int)

# Display updated column
print(df["Dowry_Exchanged"].head())
```

```python
df.Parental_Approval = df.Parental_Approval.map({"No":0 , "Yes":1,"Partial":2 })
df.Urban_Rural = df.Urban_Rural.map({"Urban":0 , "Rural":1 })
df.Marital_Satisfaction = df.Marital_Satisfaction.map({"Low":0 , "High":1,"Medium":2 })
df.Divorce_Status = df.Divorce_Status.map({"No":0 , "Yes":1 })
df.Spouse_Working = df.Spouse_Working.map({"No":0 , "Yes":1 })
df['Inter-Caste'] = df['Inter-Caste'].map({"No":0 , "Yes":1 })
df['Inter-Religion'] = df['Inter-Religion'].map({"No":0 , "Yes":1 })
```

```python
df.Income_Level = df.Income_Level.map({"Low":0 , "Middle":1,"High":2 })
```

```python
df
```

```python
x = df.drop(columns=['Marriage_Type'])  # Independent Variables
y = df['Marriage_Type']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test  = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(max_depth=2)
treemodel.fit(x_train , y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (20,15))
tree.plot_tree(treemodel, filled = True)
plt.tight_layout()
```

```python
tree_model = DecisionTreeClassifier(max_depth = 2)
tree_model.fit(x_train , y_train)
```

```python
tree.plot_tree(tree_model , filled = True)
plt.tight_layout()
```

```python
y1 = treemodel.predict(x_test)
y2 = tree_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
print(classification_report(y1 , y_test))
```

```python
print(classification_report(y2 , y_test))
```

```python
print(accuracy_score(y1 , y_test))
print(accuracy_score(y2 , y_test))
```
