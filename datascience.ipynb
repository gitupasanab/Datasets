```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Disease_symptom_and_patient_profile_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Disease"] = df["Disease"].astype("category").cat.codes
df["Fever"] = df["Fever"].map({"No": 0, "Yes": 1})
df["Cough"] = df["Cough"].map({"No": 0, "Yes": 1})
df["Fatigue"] = df["Fatigue"].map({"No": 0, "Yes": 1})
df["Difficulty Breathing"] = df["Difficulty Breathing"].map({"No": 0, "Yes": 1})
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df["Blood Pressure"] = df["Blood Pressure"].map({"Low": 0, "Normal": 1, "High": 2})
df["Cholesterol Level"] = df["Cholesterol Level"].map({"Low": 0, "Normal": 1, "High": 2})
df["Outcome Variable"] = df["Outcome Variable"].map({"Negative": 0, "Positive": 1})
```

```python
x=df.iloc[:,:-1]
y=df['Outcome Variable']
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
x_train , x_test , y_train , y_test = train_test_split(x ,y, test_size = .25 , random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
classifier  = LogisticRegression()
```

```python
parameter = {"penalty":["l1" ,"l2" , "elasticnet"] , "C":[1,2,3,4,56,5,7,9,30,54,60],"max_iter":[100,200,300,400,500]}
```

```python
classifier_reg = GridSearchCV(classifier , param_grid = parameter , scoring = "accuracy" , cv = 10)
```

```python
classifier_reg.fit(x_train , y_train)
```

```python
print(classifier_reg.best_params_)
```

```python
classifier_reg.best_score_
```

```python
y_pred = classifier_reg.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score , classification_report
```

```python
accuracy_score(y_pred , y_test)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Disease_symptom_and_patient_profile_dataset.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df["Disease"] = df["Disease"].astype("category").cat.codes
df["Fever"] = df["Fever"].map({"No": 0, "Yes": 1})
df["Cough"] = df["Cough"].map({"No": 0, "Yes": 1})
df["Fatigue"] = df["Fatigue"].map({"No": 0, "Yes": 1})
df["Difficulty Breathing"] = df["Difficulty Breathing"].map({"No": 0, "Yes": 1})
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})
df["Blood Pressure"] = df["Blood Pressure"].map({"Low": 0, "Normal": 1, "High": 2})
df["Cholesterol Level"] = df["Cholesterol Level"].map({"Low": 0, "Normal": 1, "High": 2})
df["Outcome Variable"] = df["Outcome Variable"].map({"Negative": 0, "Positive": 1})
```

```python
df
```

```python
x = df.iloc[:,:-1]
y = df['Outcome Variable']
```

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test  = train_test_split(x , y , test_size = 0.2 , random_state = 42)
```

```python
from sklearn.tree import DecisionTreeClassifier
treemodel = DecisionTreeClassifier(max_depth=2)
treemodel.fit(x_train , y_train)
```

```python
from sklearn import tree
plt.figure(figsize = (20,15))
tree.plot_tree(treemodel, filled = True)
plt.tight_layout()
```

```python
tree_model = DecisionTreeClassifier(max_depth = 2)
tree_model.fit(x_train , y_train)
```

```python
tree.plot_tree(tree_model , filled = True)
plt.tight_layout()
```

```python
y1 = treemodel.predict(x_test)
y2 = tree_model.predict(x_test)
```

```python
from sklearn.metrics import classification_report , accuracy_score
```

```python
print(classification_report(y1 , y_test))
```

```python
print(classification_report(y2 , y_test))
```

```python
print(accuracy_score(y1 , y_test))
print(accuracy_score(y2 , y_test))
```
